```python
class Animal():
 
  def __init__(self, name, color):
    self.__nome = name
    self.__cor = color
 
  @property
  def cor(self):
    return self.__cor
 
  @property
  def nome(self):
    return self.__nome
 
  @cor.setter
  def cor(self, color):
    self.__cor = color
  
  @nome.setter
  def nome(self, name):
    self.__nome = name

class Gato(Animal):
  
  def miar(self):
    print("Miau!!")

class Burro(Animal):
  
  def relinchar(self):
    print("Ióoooooooooooo")
 

 
  def __str__(self):
    return "Eu sou " + str(self.__nome)
```

```python
gatinho = Gato("Gato de Botas", "Laranja")
burrinho = Burro("Gato de Botas", "Laranja")
```

```python
gatinho.miar()
burrinho.relinchar()
```

```python
class Animal():
 
  def __init__(self, name, color):
    self.__nome = name
    self.__cor = color
 
  @property
  def cor(self):
    return self.__cor
 
  @property
  def nome(self):
    return self.__nome
 
  @cor.setter
  def cor(self, color):
    self.__cor = color
  
  @nome.setter
  def nome(self, name):
    self.__nome = name

class Gato(Animal):

  def __init__(self, name, color, ataque):
    super().__init__(name, color)
    self.__ataque = ataque
  
  def miar(self):
    print("Miau!!")

  def atacar(self):
    print("Atacar com", self.__ataque)  

class Burro(Animal):

  def relinchar(self):
    print("Ióoooooooooooo")
  
  def __str__(self):
    return "Eu sou " + str(self.__nome)
```

```python
gatinho = Gato("Gato de Botas", "Laranja", "Garras")
```

```python
gatinho.atacar()
```
