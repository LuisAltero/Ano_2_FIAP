Exercício 1 ( Orientação à objetos)

Crie um programa que armazene dados da classe Cliente. Cliete deve ter nome, email, saldo em reais e endereço.

Os clientes PJ (Pessoa Jurídica) podem ter CNPJ e site.

Os clientes PF (Pessoa Física) podem ter CPF e Instagram.

O software deve ser capaz de inserir e recuperar os dados via métodos getters e setters.

O cliente pode ainda depositar uma quantia em reais, sacar uma quantia em reais e também ver o extrato (histórico de todos os saques e depósitos).

```python
class Cliente():
  def __init__(self, nome, email, saldo, endereco):
    self.__nome = nome
    self.__email = email
    self.__saldo = saldo
    self.__endereco = endereco
    
  dict_saldo = {'Depositos' : [], 'Saques' : [], 'Saldo_atual' : []}
  lista_dep = []
  lista_saq = []

  @property
  def nome(self):
    return self.__nome
 
  @property
  def email(self):
    return self.__email

  @property
  def saldo(self):
    return self.__saldo
  
  @property
  def endereco(self):
    return self.__endereco

  @property
  def extrato(self):
    return self.dict_saldo
 
  @nome.setter
  def nome(self, nome):
    self.__nome = nome

  @email.setter
  def email(self, email):
    self.__email = email

  @endereco.setter
  def endereco(self, endereco):
    self.__endereco = endereco

  @saldo.setter
  def depositar(self, valor):
    self.__saldo += valor
    self.lista_dep.append(valor)
    self.dict_saldo['Depositos'] = self.lista_dep
    self.dict_saldo['Saldo_atual'] = self.__saldo

  @saldo.setter
  def sacar(self, valor):
    if valor > self.__saldo:
      print('Saldo insuficiente')
    else:
      self.__saldo -= valor
      self.lista_saq.append(valor)
      self.dict_saldo['Saques'] = self.lista_saq
      self.dict_saldo['Saldo_atual'] = self.__saldo


class Cliente_PF(Cliente):
  def __init__(self, nome, email, saldo, cpf, instagram, endereco):
    super().__init__(nome, email, saldo, endereco)
    self.__instagram = instagram
    self.__cpf = cpf

  @property
  def cpf(self):
    return self.__cpf

  @property
  def instagram(self):
    return self.__instagram
 
  @cpf.setter
  def cpf(self, cpf):
    self.__cpf = cpf
  
  @instagram.setter
  def instagram(self, instagram):
    self.__instagram = instagram



class Cliente_PJ(Cliente):
  def __init__(self, nome, email, saldo, cnpj, site, endereco):
    super().__init__(nome, email, saldo, endereco)
    self.__cnpj = cnpj
    self.__site = site

  @property
  def cnpj(self):
    return self.__cnpj

  @property
  def site(self):
    return self.__site
 
  @cnpj.setter
  def cnpj(self, cnpj):
    self.__cnpj = cnpj
  
  @site.setter
  def site(self, site):
    self.__site = site
```
