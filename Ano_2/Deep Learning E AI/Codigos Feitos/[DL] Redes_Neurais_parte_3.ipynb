```python
from sklearn.neural_network import MLPClassifier
import numpy as np

dataset = np.loadtxt('/home/wine.txt', delimiter=',')

entradas = dataset[:, 1:]

saidas = dataset[:,0]

redeneural = MLPClassifier (verbose=True, 
                            max_iter=10000, 
                            tol=0.000001,
                            activation='logistic',
                            learning_rate_init =0.0001)

redeneural.fit(entradas, saidas)
```

```python
redeneural.predict([[14.23,1.71,2.43,15.6,127,2.8,3.06,.28,2.29,5.64,1.04,3.92,1065]])
```
