# Aula

```python
import pandas as pd 
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv") 
df
```

```python
# Cálculo da média (aritmética) da variável price da população
df['price'].mean()
```

```python
# Cálculo da mediana da variável price da população
df['price'].median()
```

```python
df['price'].std()
```

```python
# Representação das distribuições dos gráficos: histograma
df['price'].hist()
```

```python
amostra = df.sample(3000)
```

```python
df_preco = df.agg(media_pop = pd.NamedAgg('price', 'mean'), 
                  mediana_pop = pd.NamedAgg('price', 'median'),
                  desvio_padrao_pop = pd.NamedAgg('price', 'std')).reset_index()
df_preco
```

```python
amostra_preco = amostra.agg(media = pd.NamedAgg('price', 'mean'), 
                            mediana = pd.NamedAgg('price', 'median'),
                            desvio_padrao = pd.NamedAgg('price', 'std')).reset_index()
amostra_preco
```

```python
comp_preco = pd.merge(amostra_preco, df_preco, how='outer')
comp_preco
```

```python
# Cálculo de decis e percentis
df['price'].quantile(np.arange(0,1,0.25))
```

```python
amostra['price'].quantile(np.arange(0,1,0.25))
```
