```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from scipy import stats

# População df
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv")
df.head()
```

```python
# Gráfico de dispersão - analisar as possíveis relações entre variáveis
df.plot.scatter(x='price',y='carat')
```

```python
# Coeficiente de Correlação Linear (Spearman)
df[['price','carat']].corr()
```

```python
# Matriz de correlação (linear spearman)
df[['price','carat','table']].corr()
```

```python
# Análise da relação gráfica entre price e table
df.plot.scatter(x='price',y='table')
```

```python
df.corr()
```

```python
# relação de uma reta, quando analisamos uma variável com ela mesma
df.plot.scatter(x='price',y='price')
```

```python
# relação com coeficiente de correlação negativo
df.plot.scatter(x='depth',y='table')
```

```python
# Matriz de correlação Gráfica
sn.heatmap(df.corr(), annot=True)
plt.show()
```
