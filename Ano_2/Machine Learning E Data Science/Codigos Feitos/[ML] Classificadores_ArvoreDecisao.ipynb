```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import svm

df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/dados_arvore_titanic.csv')
```

```python
# Pre processing
ageMed = df['Age'].mean()
df['Age'] = df['Age'].fillna(ageMed)

dummies_Sex = pd.get_dummies(df['Sex'])
df['SexMale'] = dummies_Sex['male']
df['SexFemale'] = dummies_Sex['female']
```

```python
# Exercício 1 - implementar a separação de base treino e teste, sendo 67,34% na base treino e o restante no teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['Pclass', 'Fare', 'SexMale']], df['Survived'],test_size=0.3266, random_state=33)
```

```python
# Modelo de árvore de decisão
mod = tree.DecisionTreeClassifier()
mod = mod.fit(x_treino, y_treino)
```

```python
# Previsão em teste
y_prev = mod.predict(x_teste)
# Acurácia retorna a porcentagem de acertos do modelo
accuracy_score(y_teste, y_prev)
```

```python
# Matriz de Confusão
pd.crosstab(y_prev,y_teste, margins=True)
```

```python
# Retorna a porcentagem de acertos da linha True positive dividido pelo (TP + FP)
precision_score(y_teste, y_prev)
```

```python
# Retorna a porcentagem de acertos da linha True positive dividido pelo (TP + FN)
recall_score(y_teste, y_prev)
```

```python
def best_metrics_values(num_comb, num_times, x, y, lib):
  import operator
  import pandas as pd
  import numpy as np
  from sklearn.model_selection import train_test_split
  from sklearn import tree
  from sklearn.metrics import accuracy_score
  from itertools import combinations
  from sklearn.metrics import recall_score
  from sklearn.metrics import precision_score
  from sklearn.ensemble import RandomForestClassifier
  from sklearn.ensemble import AdaBoostClassifier
  from sklearn import svm

  comb_cols = {}
  mod = lib
  num_tests = 0
  for number in range(1,num_comb + 1):
    pct = {}
    comb = combinations(x.columns, number)

    for i in list(comb):
      i = list(i)

      for num in range(num_times):
        X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=num)
        mod = mod.fit(X_train[i], y_train)
        y_prev = mod.predict(X_test[i])
        pct[num] = [
                    (precision_score(y_test, y_prev) + recall_score(y_test, y_prev))/2,
                    accuracy_score(y_test, y_prev), 
                    precision_score(y_test, y_prev), 
                    recall_score(y_test, y_prev)
                    ]
        max_key = max(pct.items(), key=operator.itemgetter(1))[0]
        i_str = str(i)
        comb_cols[i_str] = pct[int(max_key)]
        num_tests += 1

  max_key = max(comb_cols, key=comb_cols.get)
  accuracy = comb_cols[max_key][1]
  precision = comb_cols[max_key][2]
  recall = comb_cols[max_key][3]
  best_value = comb_cols[max_key][0]

  return print(f'\nColuna Utilizada: {max_key}\n--------------------------\nTestes Feitos: {num_tests}\n\nMelhor porcentagem obtida: {round(best_value*100, 3)}%\nAccuracy obtida: {round(accuracy*100, 3)}%\nPrecision obtida: {round(precision*100, 3)}%\nRecall obtida: {round(recall*100, 3)}%')
```
